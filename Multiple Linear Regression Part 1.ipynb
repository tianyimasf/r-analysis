### Problem 1 (50 Points)

**PART A:**

Prove that $t = \frac{\widehat\beta_1}{\widehat{s.e.}(\widehat\beta_1)}$ can be written in the form of a t-distributed random variable with $n-2$ degrees of freedom. Specifically, let $Z$ be defined as $Z = \frac{\widehat\beta_1}{\sqrt{\frac{\sigma^2}{\sum^n_{i=1}\left(x_i - \bar{x}\right)^2}}}$, and $W = \frac{(n-2)\widehat\sigma^2}{\sigma^2}$. Show that

\begin{align*}
\frac{Z}{\sqrt{W\big/(n-2)}} = \frac{\widehat\beta_1}{\widehat{s.e.}(\widehat\beta_1)} = t  ,
\end{align*}

which, by definition, means that $t \sim t(n-2)$.

**Answer**

$$\begin{align} \frac{Z}{\sqrt{W / (n-2)}} &= \frac{\widehat{\beta_1}}{\sqrt{\sigma^2(n-2)/(n-2)\widehat{\sigma}^2}} \\
                                        &= \frac{\widehat{\beta_1}}{\sigma^2 / \widehat{\sigma}^2} \\
                                        &= \frac{\widehat{\beta_1}}{\sqrt{\frac{\sigma^2}{\sum^n_{i=1}\left(x_i - \bar{x}\right)^2}}} \\
                                        &= \frac{\widehat{\beta_1}}{\widehat{s.e.}(\widehat\beta_1)} = t\end{align}$$

**PART B:**

In class, we defined the *hat* or *projection* matrix as 

$$H = X(X^TX)^{-1}X^T.$$

The goal of this question is to use the hat matrix to prove that the fitted values, $\widehat{\mathbf Y}$, and the residuals, $\widehat{\boldsymbol\varepsilon}$, are uncorrelated. We will do it in steps.

First, show that $\widehat{Y} = HY$. That is, $H$ "puts a hat on" $Y$.

**Answer**

$$\begin{align}
    HY &= X(X^TX)^{-1}X^TY \\
    &= X(X^TX)^{-1}X^T(X\widehat\beta) \text{   given ($Y = X\widehat\beta$)}\\
    &= X(X^TX)^{-1}(X^TX)\widehat\beta \\
    &= XI\widehat\beta \\
    &= X\widehat\beta \\
    &= \widehat{Y} \text{   given ($Y = X\widehat\beta + \epsilon$)}
\end{align}$$

**PART C:** 
Show that $H$ is symmetric: $H = H^T$.

**Answer**

$$\begin{align}
    H &= HI \\
    &= H(HH^T) \\
    &= H(H^TH^T) \text{ given $AB^T = A^TB^T$} \\
    &= (HH^T)H^T \\
    &= H^T \text{ Q.E.D.}
\end{align}$$

**PART D:** Show that $H(I_n - H)= 0_n$, where $0_n$ is the zero matrix of size $n \times n$.

**Answer**

$$\begin{align}
    H(I_n - H) &= H - HH \\
    &= H - (X(X^TX)^{-1}X^T)(X(X^TX)^{-1}X^T) \\
    &= H - X(X^TX)^{-1}X^TX(X^TX)^{-1}X^T \\
    &= H - X\left[(X^TX)^{-1}(X^TX)\right](X^TX)^{-1}X^T \\
    &= H - XI(X^TX)^{-1}X^T \\
    &= H - X(X^TX)^{-1}X^T \\
    &= H - H \\
    &= 0_n \text{ Q.E.D.}
\end{align}$$

**PART E:**

Stating that $\widehat{\mathbf Y}$ is uncorrelated with $\widehat{\boldsymbol\varepsilon}$ is equivalent to showing that these vectors are orthogonal. That is, we need to show that their dot product is zero:**

$$ \widehat{\mathbf Y}^T\widehat{\boldsymbol\varepsilon} = 0. $$ **Prove this result.**

**Answer**

$$\begin{align}
    \widehat{\mathbf Y}^T\widehat{\boldsymbol\varepsilon}
    &= (HY)^T\widehat{\boldsymbol\varepsilon} \\
    &= (Y^T)(H^T)(Y - X\widehat\beta) \\
    &= (Y^T)(H^TY - H^TX\widehat\beta) \\
    &= (Y^T)(HY - HX\widehat\beta) \text{ given $H = H^T$}\\
    &= (Y^T)(HY - HX\widehat\beta) \\
    &= (Y^T)(HY - H\widehat Y) \\
    &= (Y^T)(HY - HHY) \\
    &= (Y^T)(HY - HY) \text{ given $H(I_n - H)= 0_n$}\\
    &= (Y^T)0_n \\
    &= 0 \text{ Q.E.D.}\\
\end{align}$$

**PART F:** Why is this result important in the practical use of linear regression?

**Answer**

This result is important because it means that the error term doesn't contain extra information that can help with the current prediction of $Y$, which is $\widehat Y$.

### Problem 2 - Polynomial Regression via Multiple Linear Regression (50 points)

It's not too difficult to believe that some relationships between features and the response are nonlinear. Consider the following example, where the single feature $x$ and the response $y$ have a quadratic relationship of the form 

$$
Y = \frac{1}{4} - X + X^2 + \epsilon 
$$

**Part A:** Write code that samples $n=30$ data points from the true model, fits the SLR model, and plots it against the data. It should be clear from both the picture and the $R^2$-value that the SLR model is a very poor fit of the data. 

```r
# Set seed for reproducibility
set.seed(123)

# Generate x values
n <- 30
x <- seq(0, 10, length.out = n)

# Define quadratic function
quadratic_function <- function(x) {
  return(1/4 - x + x^2)
}

# Generate true y values
y_true <- quadratic_function(x)

# Add Gaussian noise
noise <- rnorm(n, mean = 0, sd = 3)  # Adjust sd for noise level
y_noisy <- y_true + noise

# Fit a linear regression model
linear_model <- lm(y_noisy ~ x)

# Get R-squared value
r_squared <- summary(linear_model)$r.squared

# Plot the data
plot(x, y_noisy, pch = 16, col = "blue", main = "Single Variable Linear Regression on Quadratic Data")

# Add the regression line
abline(linear_model, col = "green", lwd = 2)

# Add legend
legend("topleft", legend = c("Noisy Data", "Linear Fit", paste("R² =", round(r_squared, 4))),
       col = c("blue", "green", "black"), pch = c(16, NA, NA), lty = c(NA, 1, NA))
```

**Part B**: We can fit a **polynomial** model to the single-feature data be thinking of the polynomial features as features in a multiple linear regression model.  If we make the association  

$$
x_1 = x \quad \textrm{and} \quad x_2 = x^2 
$$

then we can fit a multiple linear regression of the form 

$$
\hat{y} = \hat{\beta}_0 + \hat{\beta}_1 x_1 + \hat{\beta}_2 x_2 
$$

Fit this model and print the regression coefficients to the screen. 

```r
# Fit a quadratic regression model (including x and x^2)
quadratic_model <- lm(y_noisy ~ x + I(x^2))

# Extract coefficients
coefficients <- coef(quadratic_model)
beta_0 <- round(coefficients[1], 4)
beta_1 <- round(coefficients[2], 4)
beta_2 <- round(coefficients[3], 4)

# Print model coefficients
cat("Model Coefficients:\n")
cat(paste("Beta 0:", beta_0, "\n"))
cat(paste("Beta 1:", beta_1, "\n"))
cat(paste("Beta 2:", beta_2, "\n"))
```

**Part C**: Write down the estimated MLR model in terms of the features $x_1$ and $x_2$ as well as the interpretation of the associated polynomial model in terms of the single feature $x$. Does this model seem close to the true model that the data was generated from?  

**Answer**

$\widehat y = 1.4947 - 1.4103x_1 + 1.0196x_2$. This means that when $x = 0$, $y$ is positive and equals to $1.4947$. In the original model, the $y$ value is also positive but is only $0.25$. This could potentially be explained by the noises in the generated data.

$\widehat \beta_1 = -1.4103$ which means that for small values of x, as the absolute value of $x$ increases for 1 unit, the value of $y$ decreases for $1.4103$. $\widehat \beta_2 = 1.0196$ which means that $y$ is U-shaped and increasing x initially decreases y and then increases.

**Part D**: Plot the obtained polynomial regression model against the data.  Comment on the fit.

```r
# Generate fitted values
x_fit <- seq(min(x), max(x), length.out = 100)  # More points for a smooth curve
y_fit <- predict(quadratic_model, newdata = data.frame(x = x_fit))

# Plot the data
plot(x, y_noisy, pch = 16, col = "blue", main = "Quadratic Regression on Noisy Data")

# Add fitted quadratic curve
lines(x_fit, y_fit, col = "green", lwd = 2)

# Add legend with beta values
legend("topleft", legend = c("Noisy Data", "Quadratic Fit", 
                              bquote(beta[0] == .(beta_0)), 
                              bquote(beta[1] == .(beta_1)), 
                              bquote(beta[2] == .(beta_2))),
       col = c("blue", "green", "black", "black", "black"), 
       pch = c(16, NA, NA, NA, NA), lty = c(NA, 1, NA, NA, NA))
```
